```python
# 1. Merge demo+experiment
df_new = pd.merge(df_demo, df_experiment_client, on='client_id', how='inner')
display(df_new)

# 1.1 unique id 70609
#df_new['client_id'].nunique()

# 1.2 data types
df_new.dtypes
```

```python
# 2. Data Cleaning and wrangling 

# 2.1 Map the consistancy of the name of columns 
#df_new.columns=df_new.columns.str.replace('_', ' ').str.title()

# 2.2 Drop Duplicates 
df_new.drop_duplicates(inplace=True)   

# 2.3 Check Missing values DF[Demo experiment]
df_new.isnull().sum()


# 2.4 Replace missing values 
# --> Float values with the mean value of their respective columns: 'clnt_tenure_yr', 'clnt_tenure_mnth', 'clnt_age', 'num_accts', 'float64', 'bal', 'calls_6_mnth', 'logons_6_mnth'
# --> Object (string) values with "Unknown": â€˜gendrâ€™, 'Variation', 'visitor_id', 'visit_id', 'process_step', 'object', 'date_time'  

def replace_missing_value(df_new):
    # Create a copy of the original DataFrame
    df_new2 = df_new.copy()
    
    for column in df_new2.select_dtypes(include=[np.float64]):
         df_new2 [column].fillna( df_new2 [column].mean(), inplace=True)

    for column in  df_new2 .select_dtypes(include=[object]):
         df_new2 [column].fillna("Unknown", inplace=True)
    
    display( df_new2 )

    return  df_new2 

df_new2 = replace_missing_value(df_new)
```
